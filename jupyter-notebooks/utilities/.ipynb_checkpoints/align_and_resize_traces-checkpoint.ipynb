# Filter, Align, and Resize Traces
**WARNING: This Notebook alters the Traces in the configured directory**

This script removes traces from disk that are not close to the size of the median trace. It this aligns the remaining traces and ensures that they are all of identical lengths

Alignment is accomplished using the <a href="https://github.com/pearsonkyle/Signal-Alignment">phase alignment algorithm</a>
*Credit to Pearson K. A., et al., 2019, AJ, 157, 21 (https://github.com/pearsonkyle/Signal-Alignment)*

As this notebook alters files on the disk, you should make sure you understand the assumptions accept the risk. Always back up your data before running a disk altering script. 

```python
import os, random, gc, warnings
import scipy.io
from scipy.stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm #progress bar for visualization of progress
from scipy.ndimage import shift

#Signal alignment algorithm (https://github.com/pearsonkyle/Signal-Alignment)
#Credit to Pearson K. A., et al., 2019, AJ, 157, 21
from signal_alignment import phase_align, chisqr_align


# Helper function to return directory contents sorted by datetime (oldest first)
def sorted_ls(path, sortby="mtime", require_ext=False):
    if sortby=="fsize":
        sortkey = lambda f: os.path.getsize(os.path.join(path, f))
    else:
        sortkey = lambda f: os.stat(os.path.join(path, f)).st_mtime
    file_list = list(sorted(os.listdir(path), key=sortkey))
    if require_ext != False:
        file_list = [x for x in file_list if x.endswith(require_ext)]
    return file_list

# Helper function to standardize graphs and reduce code size
def ShowTrace(title, xlabel, ylabel, ydata, fig_size=(10, 6), xdata=None, ylim=None):
    plt.figure(figsize=fig_size, dpi=80)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if ylim!= None:
        ax = plt.gca()
        ax.set_ylim([0, ylim]) 
    if xdata!= None:
        plt.plot(xdata, ydata)
    else:
        plt.plot(ydata)
    plt.show()
```

```python

I_UNDERSTAND_THAT_THIS_SCRIPT_WILL_ALTER_FILES_ON_DISK = False #Change to True to proceed

paths = ["../Final-Traces-v3/nRF52832/2018/medium-reduced/tinyaes_anechoic_10m_080618_attack/", 
        "../Final-Traces-v3/nRF52832/2018/medium-reduced/tinyaes_anechoic_10m_080618_template/"]

for path in paths:
    for i in range(3):
        filename = random.choice(sorted_ls(path, require_ext=".npy"))
        glance = np.absolute(np.load(path+filename))
        ShowTrace(path+filename, "Time (slice signal capture)", "Signal Power", glance)
```

```python
# This section eliminates traces that are not sized appropriately (invalid trace collections)
# For ease of alignment, only select the traces that are close to the size of the median trace

if(not I_UNDERSTAND_THAT_THIS_SCRIPT_WILL_ALTER_FILES_ON_DISK):
    raise("Must change this variable to true to acknowledge that data on disk will be changed")

permissible_size_difference = 0.02

# Discard outliers by size
for path in paths:
    dir_list = sorted_ls(path, sortby="fsize", require_ext=".npy")
    sizes = []
    for filename in tqdm(dir_list, desc="Calculating Mean size for " + path):
        sizes.append(os.path.getsize(path+filename))
    sizes = np.array(sizes)
    min_permissible = np.median(sizes)*(1.0-permissible_size_difference)
    max_permissible = np.median(sizes) *(1.0+permissible_size_difference)
    
    #as a sanity check, make sure that we aren't introducing bias by selecting a small set of traces
    ratio_of_good = len(sizes[(sizes>min_permissible) & (sizes<max_permissible)]) / len(dir_list)
    if (ratio_of_good < 0.9):
        raise Exception(str(100 - ratio_of_good*100)+"% of traces are bad")

    for filename in tqdm(dir_list, desc="Removing Bad Traces for "+ path):
        filesize = os.path.getsize(path+filename)
        if filesize < min_permissible or filesize > max_permissible:
            print("Deleting '"+path+filename+"' of size "+str(filesize)+" (median size: "+str(np.median(sizes))+")")
            os.remove(path+"/"+filename)
```

```python
# This section aligns traces according to the signal_alignment dependency (https://github.com/pearsonkyle/Signal-Alignment)

#****************************************************************************************************
# REQUIRED CONFIGURATION
#****************************************************************************************************
# If a known range at the beginning or end of the trace can be discarded, specify that here
trim_lower_bound = 0 #0 means don't trim beginning
trim_upper_bound = 0 #0 means don't trim end

# select 2000 samples from the middle of the trace as the region of interest (ROI)
# larger ROIs may cause performance problems in aligning traces
# if you don't know where to focus, select a 2000 sample range in the middle of the trace
# IMPORTANT: Region of interest is determined AFTER trimming, so the upper and lower ROI bounds are relative to the trimmed section
RegionOfInterest = [0, 10000]

#****************************************************************************************************

if(not I_UNDERSTAND_THAT_THIS_SCRIPT_WILL_ALTER_FILES_ON_DISK):
    raise("Must change this variable to true to acknowledge that data on disk will be changed")

# Create a template file using the smallest trace from either sample set and align all traces to it
# At this point we have discarded traces that are outside an acceptable size range, so smallest trace is close in size to other traces
template_trace = []
for path in paths:
    template_filename = sorted_ls(path, sortby="fsize", require_ext=".npy")[0] 
    t = np.absolute(np.load(path+template_filename))
    if len(template_trace) == 0 or len(t) < len(template_trace):
        template_trace = t

# Apply the trimming now
template_trace = template_trace if (trim_upper_bound == 0) else template_trace[:trim_upper_bound]
template_trace = template_trace if (trim_lower_bound == 0) else template_trace[trim_lower_bound:]

# The signal_alignment dependency (https://github.com/pearsonkyle/Signal-Alignment) gives a warning due to a future deprecation
# Since we don't own this code, we can't modify it and will ignore this warning here
warnings.filterwarnings("ignore", category=FutureWarning)

# Align every trace to the selected template trace
# Credit to Pearson K. A., et al., 2019, AJ, 157, 21 (https://github.com/pearsonkyle/Signal-Alignment)
for path in paths:
    dir_list = sorted_ls(path, require_ext=".npy")
    for filename in tqdm(dir_list, desc="Aligning Traces for "+path):
        trace = np.absolute(np.load(path+filename))
        trace = trace if (trim_lower_bound == 0) else trace[trim_lower_bound:]
        trace = trace if (trim_upper_bound == 0) else trace[:trim_upper_bound]
        trace = trace[:len(template_trace)] #Ensure traces are all equal lengths
        s = phase_align(template_trace, trace, RegionOfInterest)
        np.save(path+filename, shift(trace,s,mode='nearest')) #save aligned trace
        del trace #free memory for garbage collection
```

```python
# This section aligns traces using a much more processing intensive method
# it shifts traces a single data point at a time to find the best alignment

if(not I_UNDERSTAND_THAT_THIS_SCRIPT_WILL_ALTER_FILES_ON_DISK):
    raise("Must change this variable to true to acknowledge that data on disk will be changed")

# If a known range at the beginning or end of the trace can be discarded, specify that here
trim_lower_bound = 0 #0 means don't trim beginning
trim_upper_bound = 0 #0 means don't trim end

max_trace_misalignment = 0.05

# Create a template file using the smallest trace from either sample set and align all traces to it
# At this point we have discarded traces that are outside an acceptable size range, so smallest trace is close in size to other traces
template_trace = []
for path in paths:
    template_filename = sorted_ls(path, sortby="fsize", require_ext=".npy")[0] 
    t = np.absolute(np.load(path+template_filename))
    if len(template_trace) == 0 or len(t) < len(template_trace):
        template_trace = t

# Apply the trimming now
template_trace = template_trace if (trim_upper_bound == 0) else template_trace[:trim_upper_bound]
template_trace = template_trace if (trim_lower_bound == 0) else template_trace[trim_lower_bound:]


#start by lopping off the first part of the template. Allows us to slide the other traces in a single direction for alignment
new_template = template_trace[int(len(template_trace) * max_trace_misalignment):]

print(int(len(template_trace) * max_trace_misalignment))


#****************************************************
# this takes a while, so only do it when important
#Make sure that traces are all aligned
def align_trace(template, trace):
    lowest_p_test = 1
    best_candidate_trace = None
    lowest_mean_variance = 9999999

    for shift in range(int(len(template_trace) * max_trace_misalignment * 2)):
        shifted = trace[shift:]
        
        size = min(len(shifted), len(template))
        two_tail_p_test = pearsonr(template[0:size], shifted[0:size])[1] 
        mean_variance = np.mean(np.var([template[0:size], shifted[0:size]],0))
        
        #Calculate the probability that the data is uncorrelated
        #Misaligned data will have a lower correlation
        #2tail p-test gives the probability that the data sets are unrelated
        #we want to find alignment that gives the LOWEST p-test value
        #if(two_tail_p_test < lowest_p_test):
        #    best_candidate_trace = shifted[0:size]
        #    lowest_p_test = two_tail_p_test
        if(mean_variance < lowest_mean_variance):
            lowest_mean_variance = mean_variance
            best_candidate_trace = shifted[0:size]
        
    #print("   shifted off "+str(len(trace) - len(best_candidate_trace))+" data points")
    #print("Lowest variance: "+str(lowest_mean_variance))
    return best_candidate_trace

completed = {}
for path in paths:
    dir_list = sorted_ls(path, require_ext=".npy")
    for filename in tqdm(dir_list, desc="Aligning Traces"):
        completed[filename] = {}
        trace = np.absolute(np.load(path+filename))
        new_trace = align_trace(new_template, trace)
        completed[filename] = len(new_trace)
        del trace #free memory for garbage collection
        np.save(path+filename, new_trace) #save aligned trace
        del new_trace #free memory for garbage collection
```

```python
# Now that things are aligned, we need to again ensure that they are the same length

# Create a template file using the smallest trace from either sample set and align all traces to it
# At this point we have discarded traces that are outside an acceptable size range, so smallest trace is close in size to other traces
template_length = 0
for path in paths:
    template_filename = sorted_ls(path, sortby="fsize", require_ext=".npy")[0] 
    t = np.load(path+template_filename)
    if template_length == 0 or len(t) < template_length:
        template_length = len(t)

for path in paths:
    dir_list = sorted_ls(path, require_ext=".npy")
    for filename in tqdm(dir_list, desc="Resizing Traces"):
        trace = np.load(path+filename)
        np.save(path+filename, trace[:template_length]) #save aligned trace
        del trace #free memory for garbage collection
```
